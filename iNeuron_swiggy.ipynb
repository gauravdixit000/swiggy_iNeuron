Affordable_shops

```python
df_affordable = df[(df['Cost_for_Two(\'in rupees\')']<=500) & (df['Rating']>=4.0)]
df_affordable
```

```python
df_affordable_restaurants = df_affordable.groupby(['Shop_Name','Rating'])["Cost_for_Two('in rupees')"].agg(np.mean)
df_affordable_restaurants
```

```python
df_affordable_restaurants = df_affordable_restaurants.reset_index()
```

```python
df_affordable_restaurants.sort_values(by="Cost_for_Two('in rupees')" , ascending = True , inplace=True)
df_affordable_restaurants
```

```python
sns.scatterplot(x=df_affordable_restaurants['Cost_for_Two(\'in rupees\')'],y=df_affordable_restaurants['Rating'],palette='rainbow')
```

```python
plt.figure(figsize=(18,7))
sns.barplot(x = df_affordable_restaurants['Shop_Name'], y = df_affordable_restaurants['Cost_for_Two(\'in rupees\')'],data = df_affordable_restaurants,palette='rainbow')
plt.title('Affordable/Under Budget and Highest rated restaurants in bangalore',fontdict={'fontsize':14,'fontweight':'bold','fontstyle':'italic'})
plt.xlabel('Shop_Name',fontsize=10,fontweight='bold')
plt.ylabel('Approx.Cost for two(in rupee)',fontsize=10,fontweight='bold')
plt.xticks(rotation = 90)
plt.show()
```

```python
df['Cuisine'].unique()
```

```python
freq_dict = {}
for i in df['Cuisine'].unique():
    Cuisine_list = i.split(',')
    for Cuisine in Cuisine_list:
        Cuisine = Cuisine.lstrip(' ')
        if Cuisine in freq_dict:
            freq_dict[Cuisine] = freq_dict[Cuisine] + 1
        else:
            freq_dict[Cuisine] = 1
print(freq_dict)
print()
print('Total Records: \t', len(freq_dict))
```

```python
dfnew1 = pd.DataFrame({'Cuisine': ((i) for i in freq_dict),'No.of shops':((freq_dict[i])for i in freq_dict) })
dfnew1
```

```python
list1 = [0.2 for i in range(48)]
list1
```

```python
plt.figure(figsize=(20,20))
plt.pie(dfnew1['No.of shops'],labels=dfnew1['Cuisine'],explode=list1,autopct="%1.1f%%",rotatelabels=True,shadow=True)
plt.title("Bangalore City",fontdict={'fontsize':25,'color':'purple'})
plt.show()
```
